```python
correlation_matrix = data[['visit_per_week', 'avg_time_in_gym', 'abonoment_type']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Key Features")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca=PCA(n_components=0.9)

pca_data=pca.fit_transform(data)
print(pca_data.shape)
```

```python
#so to capture 80 percent var 5 feautures are needed and to capture 90 percent  7 features are needed
```

```python
x=range(1,pca_data.shape[1]+1)
plt.plot(x,pca.explained_variance_ratio_)
# plt.xticks(range(1,len(data.columns)+1))
plt.ylabel("Variability")
plt.xlabel("Features")
plt.title("Feature vs Portion in Variance by PCA")
plt.show()
```

```python
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# Step 4: Plot cumulative variance
plt.figure(figsize=(10, 6))
plt.plot(cumulative_variance)
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance by Number of Principal Components')
plt.grid(True)
plt.show()
```

```python
pca_0=pca_data.T[0]
pca_1=pca_data.T[1]
pca_two=pd.DataFrame({'pca_0':pca_0,'pca_1':pca_1})
pca_two['abonoment_type']=target
```

```python
pca_two
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Scatter plot for Standard and Premium
standard = pca_two[pca_two['abonoment_type'] == 0]
premium = pca_two[pca_two['abonoment_type'] == 1]
plt.scatter(standard['pca_0'], standard['pca_1'], color='blue', label='Standard')
plt.scatter(premium['pca_0'], premium['pca_1'], color='orange', label='Premium')
plt.xlabel('PCA Component 0')
plt.ylabel('PCA Component 1')
plt.legend(title='Abonnement Type')
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)

loadings_table = []
for i, (loading_vector, feature) in enumerate(zip(loadings, data.columns)):
    loadings_table.append({
        'Feature': feature,
        'PCA 0 Loading': loading_vector[0],
        'PCA 1 Loading': loading_vector[1]
    })
    plt.arrow(0, 0, loading_vector[0], loading_vector[1], color='red')
    # plt.text(loading_vector[0] * 1.5, loading_vector[1] * 1.5, feature, color='green', ha='center', va='center')
loadings_df = pd.DataFrame(loadings_table)
print(loadings_df.to_string(index=False))

# Show the plot
plt.show()
```
